```python
%matplotlib inline
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import time
import datetime
from datetime import date

data_s=pd.read_csv("sessionlogs.csv")
data_s['time']=pd.to_datetime(data_s['time'],dayfirst=True)
print(data_s.head(5))
```

The plot below shows the number of users login to system. The highest number is 265, the average is around 24, and the smallest number is 1. The pattern is extreme so it could be unusual user login.

```python
plt.figure(figsize=(30.0, 10.0))
sns.countplot(x="personid", data=data_s.sort_values(by='personid'))
plt.savefig('personid_login.png')
plt.show()
id_times= data_s['time'].groupby(data_s['personid']).count()#可列出每個ID登入幾次
```

```python
#把登入日期拆解出來
data_s['month_time'] = pd.DatetimeIndex(data_s['time']).month
data_s['day_time'] = pd.DatetimeIndex(data_s['time']).day
data_s['year_time'] = pd.DatetimeIndex(data_s['time']).year
```

```python
#照年份分登入日期
data_s2016=data_s[data_s['year_time']==2016]
data_s2017=data_s[data_s['year_time']==2017]
```

```python
#哪個月份登入人數最多
sns.countplot(x="month_time", data=data_s.sort_values(by='month_time'))
plt.savefig('month_id.png')
plt.show()
```
